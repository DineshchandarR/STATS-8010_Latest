### Sampling Distributions and their probabilty
If one draws random samples of size n from a population with mean m and variance sd2, then as n gets large, the probability distribution of the sample mean is approximately normal with mean m and variance sd2/n. The approximation improves as n increases.

Sample V = (SDNormal)**2/n , where n is no.of sample

Sample sd = (SDNormal)/np.sqrt(n) , where n is no.of sample

Note :
 sample averages from samples of all sizes from a normal population are exactly normal.
 sample sizes of 30 or larger are sufficient for the central limit theorem to take effect, regardless of the shape of the original distribution.

```python
M = 21
n = 100 #sample size
Normal_sd = 15
sample_sd = Normal_sd/(np.sqrt(n))
sample_v = (Normal_sd**2)/n   
a = 19 #a is the point at which we are to find area under the probabilty curve
Z_val = (a - M)/sample_sd 

Z_val
```
